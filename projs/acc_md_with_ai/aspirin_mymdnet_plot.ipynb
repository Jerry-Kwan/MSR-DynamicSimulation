```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = 'data/aspirin/et-md17/'
csv_file = os.path.join(csv_path, 'metrics.csv')
```

```python
df = pd.read_csv(csv_file)
df.head(15)
```

```python
len(df)
```

```python
def plot_graph(train_loss, val_loss, title):
    """Draw curve."""
    x = np.arange(len(train_loss)) + 1

    plt.figure(dpi=150)

    plt.plot(x, train_loss, 'b--', label='Train')
    plt.plot(x, val_loss, 'r--', label='Validate')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(loc='best')

    plt.show()
```

```python
train_loss = np.array(df['train_loss'])
val_loss = np.array(df['val_loss'])
train_loss_y = np.array(df['train_loss_y'])
val_loss_y = np.array(df['val_loss_y'])
train_loss_dy = np.array(df['train_loss_dy'])
val_loss_dy = np.array(df['val_loss_dy'])

N = 200
B = 5
```

```python
plot_graph(train_loss[B:N], val_loss[B:N], 'Total Loss Curve')
```

```python
plot_graph(train_loss_y[B:N], val_loss_y[B:N], 'Energy Loss Curve')
```

```python
plot_graph(train_loss_dy[B:N], val_loss_dy[B:N], 'Force Loss Curve')
```
