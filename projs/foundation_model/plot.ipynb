```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = 'data/rslt'

csv_part_1 = os.path.join(data_path, 'checkpoints_u0_part_1/u0/results/results_.csv')
csv_part_2 = os.path.join(data_path, 'checkpoints_u0_part_2/u0/results/results_.csv')
csv_all_2 = os.path.join(data_path, 'checkpoints_u0_all_2/u0/results/results_.csv')
```

```python
df_part_1 = pd.read_csv(csv_part_1)
df_part_2 = pd.read_csv(csv_part_2)
df_all_2 = pd.read_csv(csv_all_2)
```

```python
print(len(df_part_1))
df_part_1.head()
```

```python
print(len(df_part_2))
df_part_2.head()
```

```python
print(len(df_all_2))
df_all_2.head()
```

```python
valid_loss_part_1 = df_part_1['valid_loss']
valid_loss_part_2 = df_part_2['valid_loss']
valid_loss_all_2 = df_all_2['valid_loss']
```

```python
B, E = 5, 500
valid_loss_part_1 = valid_loss_part_1[B:E]
valid_loss_part_2 = valid_loss_part_2[B:E]
valid_loss_all_2 = valid_loss_all_2[B:E]
len(valid_loss_part_1)
```

```python
def plot_graph(loss1, loss2, loss3, title, label):
    """Draw curve."""
    x = np.arange(len(loss1)) + 1

    plt.figure(dpi=180)

    plt.plot(x, loss1, 'b--', label=label[0])
    plt.plot(x, loss2, 'r--', label=label[1])
    plt.plot(x, loss3, 'g--', label=label[2])
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MAE)')
    plt.legend(loc='best')

    plt.show()
```

```python
title = 'Valid Loss (MAE)'
label = ['Old+Part', 'New+Part', 'New+ALL']
plot_graph(valid_loss_part_1, valid_loss_part_2, valid_loss_all_2, title=title, label=label)
```
